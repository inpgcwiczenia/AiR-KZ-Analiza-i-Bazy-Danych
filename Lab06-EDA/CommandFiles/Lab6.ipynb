```python
# Set path to data
DIR = "../AnalysisData/"

#Celem ćwiczenia jest pobranie opisanie przypisanego zbioru danych zgodnie z protokołem TIER i zasadami "tidy"
```

```python
import pandas as pd

tb=pd.read_csv(f"{DIR}data.csv")
```

```python
#usuniecie pierwszej kolumny
tb1=tb.drop(['Unnamed: 0'], axis=1)
tb1
```

```python
#usuwanie wartosci NaN
tb1=tb1.dropna() 
tb1['Wiek kupującego']=tb1['Wiek kupującego'].astype(int)
tb1
```

```python
tb1=tb1.rename(columns={"Dni od zakupu": "DniOdZakupu","Wiek kupującego": "Wiek","Płeć kupującego": "Płeć"})
```

```python
import numpy as np
import matplotlib.pyplot as plt
#dni od zakupu histogram
min=tb1['DniOdZakupu'].min()
max=tb1['DniOdZakupu'].max()
bins=np.linspace(min,max)

tb1.DniOdZakupu.value_counts().plot(kind='bar', figsize=(8,4))
plt.title('Dni od zakupu')
plt.ylabel('Osoby')
plt.xlabel('Dni od zakupu');
```

```python
tb1.Marka.value_counts().plot(kind='bar',figsize=(8,4))
plt.title('Marka')
plt.ylabel('Osoby')
plt.xlabel('Marka');
```

```python
tb1.Płeć.value_counts().plot(kind='bar',figsize=(8,4))
plt.title('Płeć')
plt.ylabel('Osoby')
plt.xlabel('Płeć');
```

```python
tb1.Ocena.value_counts().plot(kind='bar', figsize=(8,4))
plt.title('Ocena')
plt.ylabel('Osoby')
plt.xlabel('Ocena');
```

```python
min=tb1['Wiek'].min()
max=tb1['Wiek'].max()
bins=np.linspace(min,max)

tb1.Wiek.value_counts().plot(kind='bar', figsize=(15,7))
plt.title('Wiek')
plt.ylabel('Osoby')
plt.xlabel('Wiek')

tb1.to_csv('../AnalysisData/data_fin.csv',index=True)
```
